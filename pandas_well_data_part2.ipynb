This is looking great.  Lets see if we can clean it up.  Lets fix the axis labels automaticaly and set the x limits to start at 0 and end at max.  Also, lets label the y axis.  and make the whole thing bigger!  
I use fig.set_size_inches(width,height)
then I use ax.locator_params(nbins=4,axis='x') to set the number of ticks on the x axis.  

```python
elems=['As','Fe','S']
fig, ax = plt.subplots(1, len(elems), sharey=True)
fig.set_size_inches(10,6)
for count,elem in enumerate(elems):
    ax[count].scatter(well_data[elem][well_data.As<10],well_data.Depth[well_data.As<10],color='b',label='<10')
    ax[count].scatter(well_data[elem][np.logical_and(well_data.As>=10,well_data.As<=50)],\
                      well_data.Depth[np.logical_and(well_data.As>=10,well_data.As<=50)],color='g',label='<10')
    ax[count].scatter(well_data[elem][well_data.As>50],well_data.Depth[well_data.As>50],color='r',label='<10')
    
    ax[count].set_xlabel(elem) #labels with the element
    ax[count].set_xlim(0,np.max(well_data[elem])) #sets the limit to the maximum
    ax[count].locator_params(nbins=4,axis='x')  #gives us 4 ticks marks on the x axies
    
ax[0].set_ylim([250,0])
```

To prove this works add two more parameters to elem.

Now as I am prepping for class I realize that it is hard to see all the points on the graph.  So i think we should make them transparent or open.  Transparencey is controlled by alpha.  alpha= and then a number.  

```python
elems=['As','Fe','S','P','Na']
fig, ax = plt.subplots(1, len(elems), sharey=True)
fig.set_size_inches(12,6)
for count,elem in enumerate(elems):
    ax[count].scatter(well_data[elem][well_data.As<10],well_data.Depth[well_data.As<10],color='b',label='<10',alpha=0.5)
    ax[count].scatter(well_data[elem][np.logical_and(well_data.As>=10,well_data.As<=50)],\
                      well_data.Depth[np.logical_and(well_data.As>=10,well_data.As<=50)],color='g',label='<10',alpha=0.5)
    ax[count].scatter(well_data[elem][well_data.As>50],well_data.Depth[well_data.As>50],color='r',label='<10',alpha=0.5)
    
    ax[count].set_xlabel(elem) #labels with the element
    ax[count].set_xlim(0,np.max(well_data[elem])) #sets the limit to the maximum
    ax[count].locator_params(nbins=4,axis='x')  #gives us 4 ticks marks on the x axies
    
ax[0].set_ylim([250,0])
```

After looking at the graph I don't like the transparency.  I think it is sort of ugly.  so instead lets try open circles instead.  To do this you use the key words

facecolor='none'  this is saying no colors filling in.  You could set this color.

edgecolors=  This sets the edgecolor instead of the whole color

```python
elems=['As','Fe','S','P','Na']
fig, ax = plt.subplots(1, len(elems), sharey=True)
fig.set_size_inches(12,6)
for count,elem in enumerate(elems):
    ax[count].scatter(well_data[elem][well_data.As<10],well_data.Depth[well_data.As<10],color='b',label='<10',facecolor='none')
    ax[count].scatter(well_data[elem][np.logical_and(well_data.As>=10,well_data.As<=50)],\
                      well_data.Depth[np.logical_and(well_data.As>=10,well_data.As<=50)],color='g',label='<10',facecolor='none')
    ax[count].scatter(well_data[elem][well_data.As>50],well_data.Depth[well_data.As>50],color='r',label='<10',facecolor='none')
    
    ax[count].set_xlabel(elem) #labels with the element
    ax[count].set_xlim(0,np.max(well_data[elem])) #sets the limit to the maximum
    ax[count].locator_params(nbins=4,axis='x')  #gives us 4 ticks marks on the x axies
    
ax[0].set_ylim([250,0])
```
