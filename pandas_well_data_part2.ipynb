Now as I am prepping for class I realize that it is hard to see all the points on the graph.  So i think we should make them transparent or open.  Transparencey is controlled by alpha.  alpha= and then a number.  

```python
j=['As','Fe','S','P','Na']
f, ax = plt.subplots(1, len(j), sharey=True)
f.set_size_inches(10,5)

for i in range(len(j)):

    ax[i].scatter(well_data[j[i]][well_data.As<10],well_data.Depth[well_data.As<10],color='b',label='<10',alpha=0.5)
    ax[i].scatter(well_data[j[i]][logical_and(well_data.As>=10,well_data.As<=50)],well_data.Depth[logical_and(well_data.As>=10,well_data.As<=50)],color='g',label='10-50',alpha=0.5)
    ax[i].scatter(well_data[j[i]][well_data.As>50],well_data.Depth[well_data.As>50],color='r',label='>50',alpha=0.5)
    ax[i].set_xlabel(j[i])
    ax[i].set_xlim([0,well_data[j[i]].max()])
    ax[i].xaxis.major.locator.set_params(nbins=4) 

ax[0].set_ylim([250,0])
ax[0].set_ylabel('Depth ft')
```

After looking at the graph I don't like the transparency.  I think it is sort of ugly.  so instead lets try open circles instead.  To do this you use the key words

facecolor='none'  this is saying no colors filling in.  You could set this color.

edgecolors=  This sets the edgecolor instead of the whole color

```python
j=['As','Fe','S','P','Na']
f, ax = plt.subplots(1, len(j), sharey=True)
f.set_size_inches(10,5)

for i in range(len(j)):

    ax[i].scatter(well_data[j[i]][well_data.As<10],well_data.Depth[well_data.As<10],edgecolors='b',label='<10',facecolor='none')
    ax[i].scatter(well_data[j[i]][logical_and(well_data.As>=10,well_data.As<=50)],well_data.Depth[logical_and(well_data.As>=10,well_data.As<=50)],edgecolors='g',label='10-50',facecolor='none')
    ax[i].scatter(well_data[j[i]][well_data.As>50],well_data.Depth[well_data.As>50],edgecolors='r',label='>50',facecolor='none')
    ax[i].set_xlabel(j[i])
    ax[i].set_xlim([0,well_data[j[i]].max()])
    ax[i].xaxis.major.locator.set_params(nbins=4) 

ax[0].set_ylim([250,0])
ax[0].set_ylabel('Depth ft')
```

That appears nicer!
